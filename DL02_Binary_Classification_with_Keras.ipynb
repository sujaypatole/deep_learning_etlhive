```python
from warnings import filterwarnings
filterwarnings("ignore")
```

### Step 1 - Data Ingestion

```python
url1 = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/refs/heads/main/train_loan.csv"
print(url1)
```

```python
import pandas as pd
df = pd.read_csv(url1)
df.head()
```

### Target Feature - Loan Status
1. 0 -> Loan Rejected
2. 1 -> Loan Approved

### Step 2 - Perform basic data quality check

```python
df.info()
```

```python
# Check for null values in the data frame
df.isna().sum()
```

```python
# Check for duplicate rows in the data frame
df.duplicated().sum()
```

### Step 3 - Separate X and Y (loan_status)

```python
X = df.drop(columns=["id", "loan_status"])
X.head()
```

```python
Y = df[["loan_status"]]
Y.head()
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind="bar")
```

### Step 4 - Apply Preprocessing on X

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
# Start Preprocessing
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first")
)
```

```python
pre = ColumnTransformer(
    [
        ("num", num_pipe, con),
        ("cat", cat_pipe, cat)
    ]
).set_output(transform="pandas")
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

### Step 5 - Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.20, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6 - Build a neural network

image.png

image.png

```python
xtrain.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
from keras.callbacks import EarlyStopping
```

```python
model = Sequential(
    [
        Input(shape=(xtrain.shape[1],)),
        Dense(32, activation="relu"),
        Dropout(0.3),
        Dense(16, activation="relu"),
        Dropout(0.3),
        Dense(1, activation="sigmoid")
    ]
)
```

```python
model.summary
```

```python
model.summary() #Always use this
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
es = EarlyStopping(patience=10, restore_best_weights=True)
```

```python
hist = model.fit(xtrain, ytrain, validation_split=0.2, epochs=100, callbacks=[es])
```

### Show the learning curve

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.plot(hist.history["loss"], label="train loss")
plt.plot(hist.history["val_loss"], label="validation loss")
plt.xlabel("Epochs")
plt.ylabel("Binary Crossentropy Loss")
plt.title("Learning Curve")
plt.show()
```

### Step 7 - Model Evaluation

```python
model.evaluate(xtrain, ytrain)
```

```python
model.evaluate(xtest, ytest)
```

### Out of Sample Prediction

```python
url2 = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/refs/heads/main/test_loan.csv"
print(url2)
```

```python
xnew = pd.read_csv(url2)
xnew.head()
```

```python
pre
```

```python
xnew_pre = pre.transform(xnew)
xnew_pre.head()
```

```python
yprobs = model.predict(xnew_pre)
yprobs[0:5]
```

```python
preds = [1 if prob >= 0.5 else 0 for prob in yprobs]
preds[0:5]
```

```python
res = xnew[["id"]]
res["loan_status"] = preds
res["prob"] = yprobs
```

```python
res = res.round(4)
```
